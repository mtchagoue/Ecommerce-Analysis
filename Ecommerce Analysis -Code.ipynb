EDA = Eploratory Data Analysis

```python
#what is the sales trend?
data['order_date'].min()
```

```python
data['order_date'].max()
```

```python
data['month_year'] = data['order_date'].apply(lambda x: x.strftime('%Y -%m'))
```

```python
data['month_year']
```

```python
#this sale trend is really tricky
data.groupby('month_year').sum()['sales'].reset_index()
```

```python
#
data_monthly_trend = data.groupby('month_year').sum()['sales'].reset_index()
```

```python
data.columns
```

```python
#quantity trend
data_qty_trend = data.groupby('month_year').sum()['quantity'].reset_index()
data_qty_trend
```

```python
plt.figure(figsize=(15,6))
plt.plot(data_monthly_trend['month_year'], data_monthly_trend['sales'])
plt.xticks(rotation='vertical', size=8)
plt.show()
```
